```python
import pandas as pd
df = pd.read_csv('hw_work.csv')
```

```python
# 1
print("Number of columns:", df.shape[1])
print("Number of rows:", df.shape[0])

missing_values_column = df.columns[df.isnull().any()].tolist()
print("Columns with missing values:", missing_values_column)

print("Memory usage:")
print(df.memory_usage(deep=True).sum(), "bytes")
```

```python
# 2
desc_account_length = df['Account length'].describe()
print("Account length statistics:")
print(desc_account_length)

print("Highest value:", desc_account_length['max'])
print("Lowest value:", desc_account_length['min'])
print("Average value:", desc_account_length['mean'])
```

```python
# 3
print("All columns statistics:")
print(df.describe())
```

```python
# 4
print("First 5 rows:")
print(df.head())

print("Last 5 rows:")
print(df.tail())
```

```python
# 5 - 6
print("State column values:")
print(df['State'])
print("State and International plan values:")
print(df[['State', 'International plan']])
```

```python
# 7 - 8
print("Rows where International plan is 'Yes':")
print(df[df['International plan'] == 'Yes'])

print("Rows where International plan is 'Yes' and Churn is True:")
print(df[(df['International plan'] == 'Yes') & (df['Churn'] == True)])
```

```python
# 9 - 10
print("Row at index 14:")
print(df.iloc[14])

print("Rows from index 14 onwards:")
print(df.iloc[14:])
```

```python
# 11
df_with_state_index = df.set_index('State')
print("DataFrame with State as index:")
print(df_with_state_index)
```

```python
# 12 - 13
df_without_area_code = df.drop(columns=['Area code'])
print("DataFrame without Area code column:")
print(df_without_area_code)

df_without_row = df.drop(index=3328)
print("DataFrame without row 3328:")
print(df_without_row)
```

```python
# 14
df['total_min'] = df['Total night minutes'] + df['Total intl minutes']
print("DataFrame with new column 'total_min':")
print(df)

df['average_min'] = df['total_min'] / 2
print("DataFrame with new column 'average_min':")
print(df)
```

```python
# 15
new_table = pd.concat([df.iloc[[0]], df.iloc[[-1]]], ignore_index=True)
print("New table with first and last rows:")
print(new_table)
```

```python
# 16

new_row = {'State': 'NY', 'Account length': 100, 'Churn': False}
new_row_df = pd.DataFrame([new_row])

df_with_new_row = pd.concat([df, new_row_df], ignore_index=True)
print("DataFrame with new row:")
print(df_with_new_row)
```

```python
# 17
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

df_new = df.drop(columns=['A'])
print("Original DataFrame after drop without inplace:")
print(df)

df.drop(columns=['A'], inplace=True)
print("\nOriginal DataFrame after drop with inplace:")
print(df)
```

```python
# 18
df = pd.DataFrame({'State': ['CA', 'NY', 'TX'], 'Population': [40000000, 20000000, 30000000]})

df_with_index = df.set_index('State')
print("DataFrame with 'State' as index:")
print(df_with_index)

df_reset = df_with_index.reset_index()
print("\nDataFrame after resetting index:")
print(df_reset)
```

```python
# 19
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

print("Sum of each column:")
print(df.sum(axis=0))

print("\nSum of each row:")
print(df.sum(axis=1))
```

```python
# 20
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [5, 6, 7],
    'C': [8, 9, 10]
})

print("Condition A > 1 AND B > 4:")
print(df[(df['A'] > 1) & (df['B'] > 4)])

print("\nCondition A > 2 OR B > 6:")
print(df[(df['A'] > 2) | (df['B'] > 6)])

print("\nCondition NOT A > 2:")
print(df[~(df['A'] > 2)])

print("\nCondition A is in [1, 3]:")
print(df[df['A'].isin([1, 3])])

print("\nCondition A between 2 and 3:")
print(df[df['A'].between(2, 3)])
```
